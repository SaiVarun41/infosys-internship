```python
# Statistical Methods to Identify Outliers
import numpy as np
# Z-score Method: Identify outliers with a Z-score greater than 3 or less than -3
z_scores = np.abs(zscore(df.select_dtypes(include=[np.number])))  # Only numeric columns
outliers_z = (z_scores > 3).all(axis=1)  # Rows with any Z-score > 3
```

```python
# Filter only numeric columns for IQR calculation
numeric_df = df.select_dtypes(include=[np.number])

# Calculate Q1 and Q3 for the numeric columns
Q1 = numeric_df.quantile(0.25)
Q3 = numeric_df.quantile(0.75)
IQR = Q3 - Q1

# Identify outliers based on IQR
outliers_iqr = ((numeric_df < (Q1 - 1.5 * IQR)) | (numeric_df > (Q3 + 1.5 * IQR)))

# Display rows with outliers
df_outliers_iqr = df[outliers_iqr.any(axis=1)]
print("\nRows with IQR outliers:\n", df_outliers_iqr)
```

```python
def boxplot(stock,company):
   """
   plot the box plot to the adjusted close price

   parameters
   -----------
   stock:DataFrame
     stock dataframe
   company:str
      company name.

   Returns
   --------
   None.
   """
   plt.figure(figsize=(8,6))
   plt.boxplot(stock[(company, 'Adj Close')])  # Access using MultiIndex
   plt.ylabel('price($)')
   plt.title(f'{company} Box Plot of Adjusted Close Price',fontsize=15)
   plt.show()

for company,stock in stock_list.items():
    print(boxplot(stock,company),'\n')
    plt.show()
```

```python
def enhanced_scatter_matrix(stock, company):
    """
    Creates an enhanced scatter matrix with improved styling and correlations

    Parameters
    ----------
    stock: DataFrame
        Stock's DataFrame
    company: str
        Company name
    """
    # Create the scatter matrix with enhanced styling
    fig = plt.figure(figsize=(12, 12))
    axes = pd.plotting.scatter_matrix(stock,
                                    figsize=(12, 12),
                                    diagonal='kde',  # Show density plots on diagonal
                                    alpha=0.5,       # Increased point transparency
                                    c='navy',        # Points color
                                    marker='o',      # Circular markers
                                    s=30)            # Marker size

    # Rotate axis labels for better readability
    for ax in axes.flatten():
        ax.xaxis.label.set_rotation(45)
        ax.yaxis.label.set_rotation(0)
        ax.tick_params(labelsize=8)

    plt.suptitle(f'{company} Stock Metrics Correlation Matrix',
                 fontsize=16,
                 y=0.95)
    plt.tight_layout()
    plt.show()

# Execute the visualization for each stock
for company, stock in stock_list.items():
    enhanced_scatter_matrix(stock, company)
```

```python
def heatmap(stock, company):
    """
    Plot the correlation heatmap.

    Parameters
    ----------
    stock: DataFrame
      Stock's DataFrame.
    company: str
      Company name.

    Returns
    -------
    None.
    """
    plt.figure(figsize=(8,8))
    # Changed 'Company' to 'company' to match the actual column name
    sns.heatmap(stock.drop(columns=['company']).corr(), cmap='GnBu', annot=True, linewidths = 1, fmt='f').set_title(f'Correlation Heatmap of {company}', fontsize=15)
    plt.xticks(rotation=30, fontsize=10)
    plt.yticks(rotation=0, fontsize=10)
    plt.show()

for company,stock in stock_list.items():
    print(heatmap(stock,company),'\n')
    plt.show()
```

```python
def autocorrelation(stock, company):
    """
    Plot the autocorrelation of stocks based on adjusted close price.

    Parameters
    ----------
    stock: DataFrame
      Stock's DataFrame.
    company: str
      Company name.

    Returns
    -------
    None
    """
    tsaplots.plot_acf(stock[(company, 'Adj Close')],lags=50)
    plt.title(f'{company} Autocorrelation of Adjusted Close Price', fontsize=13)
    plt.xlabel('Lags', fontsize=10)
    plt.ylabel('Autocorrelation', fontsize=10)
    plt.show()

for company,stock in stock_list.items():
    print(autocorrelation(stock,company),'\n')
    plt.show()
```

```python
import unittest
import pandas as pd
import numpy as np
from scipy.stats import skew, kurtosis

class TestDatasetValidation(unittest.TestCase):
  @classmethod
  def setUp(cls):
    cls.df = pd.read_csv('/content/tesla.csv')
    cls.df["Date"] = pd.to_datetime(cls.df["Date"])

  def test_dataset_loading(self):
    self.assertFalse(self.df.empty, "Dataset should not be empty")

  def test_date_column_exists(self):
    self.assertIn("Date", self.df.columns, "Date column should exist")

  def test_missing_values(self):
    self.assertEqual(self.df.isnull().sum().sum(), 0, "There should be no missing values in the dataset")

  def test_outliers_detection(self):
    numeric_columns = self.df.select_dtypes(include=[np.number]).columns
    z_scores = np.abs((self.df[numeric_columns] - self.df[numeric_columns].mean()) / self.df[numeric_columns].std())
    outliers = (z_scores > 3).sum().sum()
    self.assertGreaterEqual(outliers, 0, "There should be no outliers in the dataset")

  def test_skewness_and_kurtosis(self):
    numeric_columns = self.df.select_dtypes(include=[np.number]).columns
    skewness = self.df[numeric_columns].apply(lambda x: skew(x.dropna()))
    kurtosis_value = self.df[numeric_columns].apply(lambda x: kurtosis(x.dropna()))

    self.assertTrue(not skewness.isnull().any(), "Skewness values should not be null")
    self.assertTrue(not kurtosis_value.isnull().any(), "Kurtosis values should not be null")
```

```python
unittest.main(argv=[''], verbosity=2, exit=False)
```

```python
# Feature Engineering
# Price-based features
df['Open-Close_Spread'] = df['Open'] - df['Close']
df['High-Low_Spread'] = df['High'] - df['Low']
df['Price_Range_Ratio'] = (df['High'] - df['Low']) / df['Open']
```

```python
# Volatility features
df['Daily_Volatility'] = df['Daily_Return'].rolling(window=20).std()
df['Volume_MA5'] = df['Volume'].rolling(window=5).mean()
df['Volume_MA20'] = df['Volume'].rolling(window=20).mean()
```

```python
from sklearn.decomposition import PCA
from sklearn.preprocessing import StandardScaler, Normalizer
from tensorflow.keras import layers, Model
import tensorflow as tf

# Replace RSI and MACD with these advanced features
# ------------------------------------------
# PCA Features
# ------------------------------------------
price_features = ['Open', 'High', 'Low', 'Close', 'Volume']
pca = PCA(n_components=3)
price_pca = pca.fit_transform(df[price_features])

# Add PCA components
for i in range(price_pca.shape[1]):
    df[f'PCA_{i+1}'] = price_pca[:, i]
```

```python
# ------------------------------------------
# Standardization and Normalization
# ------------------------------------------
# Standardization
scaler = StandardScaler()
df_std = scaler.fit_transform(df[price_features])
for i, col in enumerate(price_features):
    df[f'{col}_Std'] = df_std[:, i]

# Normalization
normalizer = Normalizer()
df_norm = normalizer.fit_transform(df[price_features])
for i, col in enumerate(price_features):
    df[f'{col}_Norm'] = df_norm[:, i]
```

```python
# ------------------------------------------
# Autoencoder Features
# ------------------------------------------
# Prepare data for autoencoder
X = df[price_features].values
X_scaled = StandardScaler().fit_transform(X)

# Define autoencoder architecture
input_dim = X_scaled.shape[1]
encoding_dim = 3

# Build autoencoder
input_layer = layers.Input(shape=(input_dim,))
encoded = layers.Dense(8, activation='relu')(input_layer)
encoded = layers.Dense(encoding_dim, activation='relu')(encoded)
decoded = layers.Dense(8, activation='relu')(encoded)
decoded = layers.Dense(input_dim, activation='linear')(decoded)

# Create and train autoencoder
autoencoder = Model(input_layer, decoded)
encoder = Model(input_layer, encoded)
autoencoder.compile(optimizer='adam', loss='mse')
autoencoder.fit(X_scaled, X_scaled, epochs=50, batch_size=32, shuffle=True, verbose=0)

# Generate encoded features
encoded_features = encoder.predict(X_scaled)
for i in range(encoding_dim):
    df[f'AE_Feature_{i+1}'] = encoded_features[:, i]
```

```python
# ------------------------------------------
# Step 7: Autoencoder for Feature Extraction
# ------------------------------------------
import tensorflow as tf
from tensorflow import keras
from tensorflow.keras import layers

# Select numerical features for autoencoder
numerical_features = df.select_dtypes(include=np.number).columns.tolist()
numerical_features = [f for f in numerical_features if f not in ['hour', 'Day']]
numerical_features = [f for f in numerical_features if f != 'Log_Close']
numerical_features = numerical_features + ['Log_Close']

# Prepare data for autoencoder
ae_df = df[numerical_features].dropna()
X = ae_df.values

# Scale the data
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
X_scaled_ae = scaler.fit_transform(X)

# Define autoencoder architecture
input_dim = X_scaled_ae.shape[1]
encoding_dim = 8

# Build the model
input_layer = layers.Input(shape=(input_dim,))
encoded = layers.Dense(32, activation='relu')(input_layer)
encoded = layers.Dense(encoding_dim, activation='relu')(encoded)
decoded = layers.Dense(32, activation='relu')(encoded)
decoded = layers.Dense(input_dim, activation='linear')(decoded)

# Create and compile the model
autoencoder = keras.Model(inputs=input_layer, outputs=decoded)
encoder = keras.Model(inputs=input_layer, outputs=encoded)
autoencoder.compile(optimizer='adam', loss='mse')

# Train the autoencoder
autoencoder.fit(X_scaled_ae, X_scaled_ae,
                epochs=10,
                batch_size=256,
                shuffle=True,
                verbose=1)

# Generate encoded features
X_encoded = encoder.predict(X_scaled_ae)

# Add encoded features to dataset
encoded_cols = [f'AE_Feature_{i+1}' for i in range(encoding_dim)]
encoded_df = pd.DataFrame(X_encoded, index=ae_df.index, columns=encoded_cols)
df = df.join(encoded_df)

print("Autoencoder features created successfully!")
print("New features:", encoded_cols)
print("\nSample of autoencoder features:")
print(df[encoded_cols].head())
```

```python
# ------------------------------------------
# Step 8: Additional Time-Based Aggregations
# ------------------------------------------
# Aggregate by day and hour
hour_group = df.groupby(['Day', 'hour'])['Close'].agg(['mean', 'std', 'count']).reset_index()
hour_group = hour_group.rename(columns={
    'mean': 'Hour_Mean_Price',
    'std': 'Hour_Std_Price',
    'count': 'Hour_Count'
})

# Merge aggregated features back to main dataframe
df = df.merge(hour_group, on=['Day', 'hour'], how='left')

# ------------------------------------------
# Step 9: Final Visualization and Check
# ------------------------------------------
# Print final dataset information
print("Final Data Shape:", df.shape)
print("\nNew features added:",
      [col for col in df.columns if col.startswith(('Hour_', 'AE_'))])

# Create visualization of autoencoder features
plt.figure(figsize=(15, 5))

# Plot AE_Feature_1 distribution
plt.subplot(1, 2, 1)
plt.hist(df['AE_Feature_1'].dropna(), bins=50, alpha=0.7,
         color='blue', label='AE_Feature_1')
plt.title('Distribution of AE_Feature_1')
plt.legend()

# Plot AE_Feature_2 distribution
plt.subplot(1, 2, 2)
plt.hist(df['AE_Feature_2'].dropna(), bins=50, alpha=0.7,
         color='orange', label='AE_Feature_2')
plt.title('Distribution of AE_Feature_2')
plt.legend()

plt.tight_layout()
plt.show()

# Save final processed dataset
df.to_csv('tesla_final_processed.csv', index=False)
```
